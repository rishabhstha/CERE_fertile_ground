# Scraping Articles from News Media Sources using GoogleNews API
#Note: If Google News API ever gives you a lot of "caught a timeout" and nothing as scraped data then relaunch the kernel and try again.

#### Inporting libraries

```python
from GoogleNews import GoogleNews
from newspaper import Article
import pandas as pd
from newspaper import Article
from newspaper import Config
import nltk

import warnings
warnings.filterwarnings("ignore")
```

#### Used keywords such as 'food access Mississippi', 'food security Mississippi', 'food insecurity Mississippi', 'fertile ground Mississippi' to scrape news articles

```python

keywords=['food access Mississippi', 'food security Mississippi', 'food insecurity Mississippi', 'fertile ground Mississippi']
list=[]
for keyword in keywords:
    googlenews=GoogleNews(start='03/01/2018',end='06/20/2021')
    googlenews.search(keyword)
    result=googlenews.result()
    df=pd.DataFrame(result)

    for i in range(2,15):
        googlenews.getpage(i)
        result=googlenews.result()
        df=pd.DataFrame(result)
        
    user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'
    config = Config()
    config.browser_user_agent = user_agent
    config.request_timeout = 50

    
    for ind in df.index:
        try:
            dict={}
            article = Article(df['link'][ind],config=config)
            article.download()
            article.parse()
            article.nlp()
            dict['Date']=df['date'][ind]
            dict['Media']=df['media'][ind]
            dict['Title']=article.title
            dict['Article']=article.text
            dict['Summary']=article.summary
            dict['Keyword']=keyword
            list.append(dict)
        except:
            print("caught a timeout")
    
    googlenews.clear()

news_df=pd.DataFrame(list)

#Dropping the duplicate news articles filtered by Titles
news_df=news_df.drop_duplicates(subset='Title')

news_df.to_csv("data/google_allnews.csv")
news_df
```
